```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data_siswa = {
    'nama': ['Andi', 'Budi', 'Citra', 'Dina', 'Eka', 'Fira', 'Gilang', 'Hana', 
             'Ivan', 'Joko', 'Kiki', 'Luna', 'Maya', 'Nina', 'Omar'],
    'matematika': np.random.randint(60, 100, 15),
    'bahasa_indonesia': np.random.randint(65, 95, 15),
    'ipa': np.random.randint(55, 98, 15),
    'ips': np.random.randint(70, 95, 15)
}
```

```python
dfSiswa = pd.DataFrame(data_siswa)
```

```python
# Hitung nilai rata-rata setiap siswa dan tambahkan kolom 'rata_rata'
dfSiswa['rerata'] = (dfSiswa['matematika'] + dfSiswa['bahasa_indonesia'] + dfSiswa['ipa'] + dfSiswa['ips'])/4
```

```python
#Hitung statistik deskriptif (mean, median, std) untuk setiap mata pelajaran
import pandas as pd
import numpy as np

# Misal dfSiswa berisi kolom nilai pelajaran
mata_pelajaran = ['matematika', 'bahasa_indonesia', 'ipa', 'ips']

for pelajaran in mata_pelajaran:
    mean = dfSiswa[pelajaran].mean()
    median = dfSiswa[pelajaran].median()
    std = dfSiswa[pelajaran].std()

    print(f"Statistik untuk {pelajaran.capitalize()}:")
    print(f"  Mean   : {mean:.2f}")
    print(f"  Median : {median:.2f}")
    print(f"  Std    : {std:.2f}")
    print()
```

```python
#Filter siswa dengan rata-rata nilai â‰¥ 80
siswaLeb80 = dfSiswa[dfSiswa['rerata'] >= 80]
siswaLeb80
```

```python
#Buat histogram distribusi nilai rata-rata siswa
plt.hist(dfSiswa['rerata']);
```

```python
# Buat bar chart perbandingan nilai rata-rata per mata pelajaran
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].bar(dfSiswa['nama'], dfSiswa['matematika'])
axs[0, 0].set_title('Nilai Matematika')
axs[0, 0].tick_params(axis='x', rotation=45)

axs[0, 1].bar(dfSiswa['nama'], dfSiswa['bahasa_indonesia'])
axs[0, 1].set_title('Nilai Bahasa Indonesia')
axs[0, 1].tick_params(axis='x', rotation=45)

axs[1, 0].bar(dfSiswa['nama'], dfSiswa['ipa'])
axs[1, 0].set_title('Nilai IPA')
axs[1, 0].tick_params(axis='x', rotation=45)

axs[1, 1].bar(dfSiswa['nama'], dfSiswa['ips'])
axs[1, 1].set_title('Nilai IPS')
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
